**Question 2. Which App has maximum number of Reviews?

```python
Googleplay_data_cleaned["Reviews"] = Googleplay_data_cleaned["Reviews"].astype(int)
```

```python
max_reviews = Googleplay_data_cleaned.groupby(['App'])[['Reviews']].max()
max_review_sorted = max_reviews.sort_values('Reviews', ascending=False)
max_review_top20= max_review_sorted.head(20)
max_review_top20
```

```python
max_review_top20 = max_review_top20.reset_index()
```

```python
df=max_review_top20.set_index("App")
df
```

```python
max_review_top20
```

```python
import matplotlib.ticker as mtick
```

```python
#Maximum number of Reviews per App (Top 20)

plot = df.plot(kind='barh', legend=False, figsize=(10,10))
def reviews(x, pos):
    return '{:,.0f} M'.format(x / 1000000)

plot.xaxis.set_major_formatter(reviews)
plt.xlabel('Number of Reviews')
plt.ylabel('App')
plt.title('Maximum number of Reviews per App (Top 20)')
plt.savefig("images/MaxReviews.png")
```

## Average Category Rating Chart

```python
from pyecharts import options as opts
from pyecharts.charts import Pie
from pyecharts.charts import Bar
from pyecharts.globals import ThemeType
```

```python
# Create bar chart
bar = Bar(init_opts=opts.InitOpts(theme=ThemeType.MACARONS))
bar.add_xaxis(avg_rating['Category'].tolist())
bar.add_yaxis("", avg_rating['Rating'].tolist())
bar.set_series_opts(label_opts=opts.LabelOpts(is_show = False))
```

```python
# Add labels
bar.set_global_opts(
    title_opts=opts.TitleOpts(title="Average ratings by category"),
    yaxis_opts=opts.AxisOpts(name='Average rating', min_=3.0, max_=4.5),
    xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=90))
)
```

```python
# Render html
bar.render("CategoryRating.html")
```

## Maximum Reviews Chart

```python
bar_chart = Bar(init_opts=opts.InitOpts(theme=ThemeType.WALDEN))
bar_chart.add_xaxis(max_review_top20['App'].tolist())
bar_chart.add_yaxis("", max_review_top20['Reviews'].tolist())
bar_chart.set_series_opts(label_opts=opts.LabelOpts(is_show = False))
```

```python
bar_chart.set_global_opts(
    xaxis_opts=opts.AxisOpts(
        axislabel_opts={"rotate": 20, "font_size": 12, "font_weight": "bold"}
    ),
    yaxis_opts=opts.AxisOpts(
        axislabel_opts={"font_size": 12, "font_weight": "bold"},
        splitline_opts=opts.SplitLineOpts(is_show=True),
    ),
    title_opts=opts.TitleOpts(
        title="Maximum number of Reviews per App (Top 20)", pos_left="center"
    ),
)
```

```python
bar_chart.render("MaxReviews.html")
```
