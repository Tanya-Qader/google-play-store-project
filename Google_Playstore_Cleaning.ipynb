```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
Googleplay_data = pd.read_csv("Data/googleplaystore.csv")
Googleplay_data.head()
```

```python
UserReview_data = pd.read_csv("Data/googleplaystore_user_reviews.csv")
UserReview_data.head(20)
```

```python
#Information about the dataset (columns, rows, etc)

Googleplay_data.info()
```

```python
Googleplay_data.describe(include = 'all')
```

```python
#checking for the number of null values

Googleplay_data.isna().sum().sum()
```

```python
#checking number of null values per column

Googleplay_data.isnull().sum()
```

```python
#drop the unwanted columns 

Googleplay_data = Googleplay_data.drop(Googleplay_data.columns[[10, 11, 12]], axis=1)
Googleplay_data
```

```python
#count the number of ratings
Rating_count= Googleplay_data['Rating'].value_counts()
Rating_count
```

```python
Googleplay_data['Type'].value_counts()
```

```python
Googleplay_data_cleaned = Googleplay_data.dropna()
Googleplay_data_cleaned
```

```python
Googleplay_data_cleaned = Googleplay_data_cleaned.drop_duplicates(subset=['App'])
Googleplay_data_cleaned
```

```python
Googleplay_data_cleaned.isna().sum()
```

```python
# Frequency plot of different ratings.
g = sns.kdeplot(Googleplay_data_cleaned.Rating, color="Red", shade = True)
g.set_xlabel("Rating")
g.set_ylabel("Frequency")
plt.title('Distribution of Rating',size = 20)
```

The average rating of apps in the play store is 4, which is quiet high.

```python
# Frequency plot of of apps in different categories.
g = sns.countplot(x="Category",data=Googleplay_data_cleaned, palette = "Set1")
g.set_xticklabels(g.get_xticklabels(), rotation=90, ha="right")
g 
plt.title('Count of app in each category',size = 20)
```

Family category has the highest number of app appereances followed by Game and Tool.

**Question 1: Which App category has the highest average rating?

```python
#1.	Which App category has the highest average rating?
Average_rating = Googleplay_data_cleaned.groupby(['Category'])[['Rating']].mean()
Average_rating
```

```python
Average_rating = Average_rating.reset_index()
```

```python
Average_rating
```

```python
plt.figure(figsize = (20, 5))
avg_rating = Average_rating.groupby('Category')['Rating'].mean().round(2).sort_values(ascending = False).reset_index()

fig = sns.barplot(x = avg_rating.Category, y = avg_rating.Rating, palette = 'hls')
fig.set_xticklabels(fig.get_xticklabels(), rotation = 90)
plt.ylim(3.0, 4.5)
plt.ylabel('Average rating')
plt.title('Average ratings by category')
```

**Question 2. Which App has maximum number of Reviews?

```python
Googleplay_data_cleaned["Reviews"] = Googleplay_data_cleaned["Reviews"].astype(int)
```

```python
max_reviews = Googleplay_data_cleaned.groupby(['App'])[['Reviews']].max()
max_review_sorted = max_reviews.sort_values('Reviews', ascending=False)
max_review_top20= max_review_sorted.head(20)
max_review_top20
```

```python
max_review_top20 = max_review_top20.reset_index()
```

```python
max_review_top20
```

```python
import matplotlib.ticker as mtick
```

```python
#Maximum number of Reviews per App (Top 20)

plot = max_review_top20.plot(kind='barh', legend=False, figsize=(10,10))
def reviews(x, pos):
    return '{:,.0f} M'.format(x / 1000000)

plot.xaxis.set_major_formatter(reviews)
plt.xlabel('Number of Reviews')
plt.ylabel('App')
plt.title('Maximum number of Reviews per App (Top 20)')

```

```python
# finding the different types of apps average 
app_type= Googleplay_data_cleaned.groupby(["Type"])["Type"].count()
app_type
```

```python
app_type.plot.pie(autopct='%1.0f%%' , colors = ['green', 'blue'],)

plt.title("Types of Google Play Apps")
plt.show()
```

```python
plt.figure(figsize = (10, 5))
type = Googleplay_data_cleaned.groupby('Type')['App'].count().reset_index()

plt.bar('Type', 'App', data = type, color = ['Green', 'blue'])
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Type countplot')
plt.show()
```
