```python
#drop the unwanted columns 

Googleplay_data = Googleplay_data.drop(Googleplay_data.columns[[10, 11, 12]], axis=1)
Googleplay_data
```

```python
Googleplay_data_cleaned = Googleplay_data.dropna()
Googleplay_data_cleaned
```

```python
Googleplay_data_cleaned.isna().sum()
```

**Question 1: Which App category has the highest average rating?

```python
#1.	Which App category has the highest average rating?
Average_rating = Googleplay_data_cleaned.groupby(['Category'])[['Rating']].mean()
Average_rating_df = print(Average_rating)
Average_rating_df
```

```python
plt.figure(figsize = (20, 5))
avg_rating = Average_rating.groupby('Category')['Rating'].mean().round(2).sort_values(ascending = False).reset_index()

fig = sns.barplot(x = avg_rating.Category, y = avg_rating.Rating, palette = 'hls')
fig.set_xticklabels(fig.get_xticklabels(), rotation = 90)
plt.ylim(3.0, 4.5)
plt.ylabel('Average rating')
plt.title('Average ratings by category')
```

**2. Which App has maximum number of Reviews?

```python
Googleplay_data_cleaned["Reviews"] = Googleplay_data_cleaned["Reviews"].astype(int)
```

```python
max_reviews = Googleplay_data_cleaned.groupby(['App'])[['Reviews']].max()
max_review_sorted = max_reviews.sort_values('Reviews', ascending=False)
max_review_top20= max_review_sorted.head(20)
max_review_top20
```

```python
import matplotlib.ticker as mtick
```

```python
plot = max_review_top20.plot(kind='barh', legend=False, figsize=(10,10))

def reviews(x, pos):
    return '{:,.0f} M'.format(x / 1000000)

plot.xaxis.set_major_formatter(reviews)
```

```python
plt.figure(figsize = (20,5))
fig = sns.countplot(x = max_review_top20['Reviews'].sort_values(ascending = False), order = max_review_top20['Reviews'].value_counts().index, color = 'blue')
fig.set_xticklabels(fig.get_xticklabels(), rotation = 90)
plt.title('Installs countplot')
plt.show(fig)
```
